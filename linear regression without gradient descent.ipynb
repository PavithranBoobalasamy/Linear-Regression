```python
import  numpy as np
x=np.arange(1,100)
y=np.arange(1,100)
print(x)
print("***********************************************************************************")
print(y)


x_sum=np.sum(x)
x_mean=x_sum/len(x)
y_sum=np.sum(y)
y_mean=y_sum/len(y)
print("***********************************************************************************")
print(x_mean,y_mean)
print("***********************************************************************************")
numer=0
denom=0
for i in range(len(x)):
    numer+=(y[i]-y_mean)*(x[i]-x_mean)
    denom+=(x[i]-x_mean)**2
m=numer/denom
print("m:",m)
print("***********************************************************************************")
c=y_mean-(m*x_mean)
print("c:",c)
print("***********************************************************************************")
y_prediction=np.dot(x,m)+c
print(y_prediction)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color="r",marker='.',linewidth=2)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.plot(x,y_prediction,color='blue',linewidth=2)



```
